### **3.5 eliminación de valores nulos y atípicos**

Primero se determinará el porcentaje de valores nulos en todas las columnas del dataset, con el fin de evaluar la magnitud de los datos faltantes y decidir cómo proceder en cada caso durante el proceso de limpieza.

```python
porcentaje_nulos = df_clientes.isna().mean()*100
print(porcentaje_nulos)
```

Al analizar el porcentaje de valores nulos en las diferentes columnas del dataset, se observa lo siguiente:

- La mayoría de las columnas presentan menos del 1% de valores faltantes, como es el caso de edad (0.34%) y promedio_gasto_comida (0.48%).

No obstante, para maximizar la cantidad de datos disponibles, se optará por imputar los valores faltantes en lugar de eliminarlos, especialmente en las columnas críticas como `edad` y `promedio_gasto_comida`.

#### **Columna edad**

El análisis de la columna edad revela la presencia de valores imposibles para la edad humana, como -5, 300 y valores nulos nan. Estos valores son evidentemente incorrectos y deben ser tratados durante el proceso de limpieza de datos para asegurar la integridad y confiabilidad del análisis.

```python
df_clientes['edad'].unique()
```

Se procede aplicar un filtro para remplazar los valores atipicos por valores nulos.

```python
# eligar el rango de edades que son posibles para la vidad humana
df_clientes['edad'] = df_clientes['edad'].apply(lambda x: x if (x is not None and 18 <= x <= 90) else np.nan)
```

Los valores nulos nan se imputaran mediante K vecinos más cercanos.




```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

# 1. Definimos las columnas numéricas a trabajar (en este caso solo 'edad')
columnas_numericas = ['edad']
df_numerico = df_clientes[columnas_numericas]

# 2. Escalamos los datos para que tengan media 0 y desviación estándar 1
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numerico)

# 3. Imputamos los valores faltantes con KNN (k=5 vecinos)
imputer = KNNImputer(n_neighbors=5)
df_imputado_scaled = imputer.fit_transform(df_scaled)

# 4. Invertimos la transformación para devolver los valores a su escala original
df_imputado = scaler.inverse_transform(df_imputado_scaled)
# 5. Reemplazamos en el DataFrame original la columna 'edad' con los valores ya imputados
df_clientes[columnas_numericas] = pd.DataFrame(df_imputado, columns=columnas_numericas)
```

#### **Columna promedio_gasto_comida**

por lo tanto se procede con la imputacion mediante K vecinos más cercanos

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

# 1. Definimos las columnas numéricas a trabajar (en este caso solo 'edad')
columnas_numericas = ['promedio_gasto_comida']
df_numerico = df_clientes[columnas_numericas]

# 2. Escalamos los datos para que tengan media 0 y desviación estándar 1
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numerico)

# 3. Imputamos los valores faltantes con KNN (k=5 vecinos)
imputer = KNNImputer(n_neighbors=5)
df_imputado_scaled = imputer.fit_transform(df_scaled)

# 4. Invertimos la transformación para devolver los valores a su escala original
df_imputado = scaler.inverse_transform(df_imputado_scaled)
# 5. Reemplazamos en el DataFrame original la columna 'edad' con los valores ya imputados
df_clientes[columnas_numericas] = pd.DataFrame(df_imputado, columns=columnas_numericas)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar nulos en edad
df_edad_licor = df_clientes.dropna(subset=["edad", "consume_licor"])

plt.figure(figsize=(8,6))
sns.boxplot(
    data=df_edad_licor,
    x="consume_licor",
    y="edad",
    palette="Set2"
)
plt.title("Relación entre consumo de licor y edad de los clientes")
plt.xlabel("Consume licor")
plt.ylabel("Edad")
plt.show()
```

```python
# Descripción estadística de la edad según consumo de licor
desc_stats = df_clientes.groupby("consume_licor")["edad"].describe()
print(desc_stats)
```

Las personas que no consumen licor tienen una edad promedio mayor (53.6 años) y se concentran entre los 40 y 70 años.
En cambio, quienes sí consumen licor son más jóvenes en promedio (46.1 años) y presentan mayor variabilidad en las edades.
